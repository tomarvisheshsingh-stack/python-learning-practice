String methods


```python
# lower()
# lower syntax : string.lower()
s = "pyTHon"
print(s)
print(id(s))
print(s.lower())
print(id(s))

s1 = s.lower()
print(s1)
print(id(s1))
```

```python
# Upper
# upper syntax : string.upper()
s = "PYthon"
print(s)
print(id(s))
print(s.upper())
print()
s1 = s.upper()
print(s1)
print(id(s1))
```

```python
# Title and Capitalize
# title syntax : string.title()
# capitalize syntax : string.capitalize()
s = 'this is python'
print(s.title())
print(s.capitalize())
```

```python
# Index and Find
# index syntex : string.index('Element', start, stop)
# Find syntax : string.find("Element", start, stop)
s = 'this is python'
print(s.index('is'))
print(s.index('is',3))
# print(s.index('si'))
print(s.find('is', 4))         
print(s.find('si'))
```

```python
# Replace
# replace syntaxx : object.replace('original_value', 'new_value')

s = 'this is python'
print(s.replace('is', 'IS'))
print(s)
print(s.replace('si', 'IS'))
```

```python
# Startswith and Endswith
# startswith syntax : string.startswith('start of element')
# endswith syntax : string.endswith('end of element')
s = 'this is python'
print(s.startswith('thi'))
print(s.startswith('Thi'))

print(s.endswith('hon'))
print(s.endswith('N'))
```

```python
# Join 
# join syntax : "separator".join(object)

l = ['python', 'java', 'php']
s = " ".join(l)
print(s)

s1, s2, s3 = "Python", "Java", "PHP"
# a = " ".join(s1, s2, s3)
a = " ".join([s1,s2,s3])
print(a)
print(type(a))

```

```python
s1, s2, s3 = "Python", "Java", "PHP"
# a = " ".join(s1, s2, s3)
a = ", ".join([s1,s2,s3])
print(a)
```

```python
# Split 
# split syntax = string.split("separator", no. of splits)      by default: separator takes space and no. of splits takes no. of spaces
# It also replaces the separator: which means the separator doesn't show up in the output
s = "python java php"
l = s.split()
a = s.split(" ", 2)
b = s.split("p")
print(b)
print(l)
print(a)
```

## list methods

### append() : used to single element at last position 
### copy() : copies the list
### clear() : empties the list
### extend() : adding multiple elements at end on list
### insert() : add single element at desired position
### pop() : deletes last element from list
### remove() : delete targeted element from the list(first occurenece)
### index() : gives indexing value of the elements in list
### count() : counts no. elements in the list 
### sort() : arrange lists in to ascending order based on ascii values
### reverse() : arrange lists in to descending order based on ascii values

```python
# append syntax : list.append(element to be added) 

l = [10, 20, 65, 5, 'python', 'ans']
l.append('java')
print(l)
```

```python
# copy syntax : list2 = list.copy()

l = [10, 20, 65, 5, 'python', 'ans']
l1 = l.copy()
print(l, l1)
print(id(l), id(l1))
```

```python
# clear syntax : list.clear()

l = [10, 20, 65, 5, 'python', 'ans']
print(l)
l.clear()
print(l)
print(id(l))

# del l         # del is used to delete from memory
print(l)
```

```python
# extend syntax : list.extend(elements)

l = [10, 20, 65, 5, 'python', 'ans']
l.extend
```

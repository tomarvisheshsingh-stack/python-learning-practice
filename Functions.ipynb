```python
def even_no(n):
    i = 2
    while i<=n:
        print(i)
        i = i + 2
n = int(input('Enter any number : '))
even_no(n)
```

```python
def odd_no(n):
    i = 1
    while i<=n:
        print(i)
        i = i + 2
n = int(input('Enter any number : '))
odd_no(n)
```

```python
def odd_for(n):
    for i in range (1, n+1, 2):
        print(i)
n = int(input('Enter any number : '))
odd_for(n)
```

## For list

```python
def max_list(l):
    max_list = []
    max_ord = 0
    for i in l:
        for j in i:
            if ord(j) > max_ord:
                max_ord = ord(j)
                max_list = i
    print(max_list)
max_list( list(input('Enter the values : ').split()))
```

```python
def count_list(l, a):
    count = 0
    for i in l:
        if i == a:
            count += 1
    print(count)
count_list( list(input('Enter the values : ').split()), input('Enter the value you want to count : ') )
```

```python
def length_list(l):
    count = 0
    for i in l:
        count += 1
    print(count)
length_list( list(input('Enter the values : ').split()) )
```

## for tuple

```python
def count_tup(t, a):
    count = 0
    for i in t:
        if i == a:
            count += 1
    print(count)
count_tup( tuple(input('Enter the values : ').split()), input('Enter the value you want to count : ') )
```

```python
def length_tup(t):
    count = 0
    for i in t:
        count += 1
    print(count)
length_tup( tuple(input('Enter the values : ').split()) )
```

## Type of arguments

### positional argument
### default positional argument
### variable-length positional argument(*args (single))
### keyword argument
### default keyword argument 
### variable length keyword argument (**args (dictionary))


```python
# positional argument

def new(x, y, z):
    print('x:', x)
    print('y:', y)
    print('z:', z)
# new()
# new(10)
# new(10, 20)
new(10, 20, 30)
```

```python
# default positional argument

def new(x = 0, y = 0, z = 0):
    print('x:', x)
    print('y:', y)
    print('z:', z)
new()
new(10)
new(10, 20)
new(10, 20, 30)
```

```python
# variable-length positional argument(*args (single))

def new(*args):
    print(args)
    print(type(args))
new()
new(1,2,3)
new(1,2,3,1,2,3,12,545,4)
```

```python
n = int(input('Enter how many no. : '))
l = []
for i in range(1, n+1):
    value =  int(input(f'Enter {i}th value: '))
    l.append(value)
print(l)
# print(*l)

def add(*n):
    sum = 0
    for i in n:
        sum = sum + i
    print(sum)
add(*l)
```

```python
n = int(input('Enter how many no. : '))
l = []
for i in range(1, n+1):
    value =  int(input(f'Enter {i}th value: '))
    l.append(value)
# print(l)
# print(*l)

def add(*n):
    print(n)
    sum = 0
    for i in n:
        for j in i:
            sum = sum + j
    print(sum)
add(l)
```

```python
# keyword positional argument
def fun(x,y,z):
    print(x,y,z)
fun(z=10, y = 5, x = 1)
# fun()
# fun(z=10)
```

```python
# default keyword positional argument 
def fun(x=0, y=0, z=0):
    print(x,y,z)
# fun()
# fun(z = 10)
fun(z=1,x=2,y=33)
```

```python
# variable length keyword positional argument (**kwargs (dictionary))
# def fun(**a):
def fun(**kwargs):
    print(kwargs)
    print(type(kwargs))
fun(z=45, x=4, y =44)
```

```python
def fun(**kwargs):
    print(kwargs)
    print(type(kwargs))
    
d = { 'name':'vish', 'age':20, 'quali':'nothing'}
# print(d)
# print(**d)
fun(**d)
fun(x=10)
# x = 10
# fun(x)         incorrect way
```

```python
def fun(**kwargs):
    for i in kwargs:
        print(i, d[i])
    
d = { 'name':'vish', 'age':20, 'quali':'nothing'}
fun(**d)
```

```python
def fun(**kwargs):
    for i,j in kwargs.items():
        print(i,j)
d = { 'name':'vish', 'age':20, 'quali':'nothing'}
fun(**d)
```

```python
def fun(**kwargs):
    for i in kwargs.items():
        print(i)
d = { 'name':'vish', 'age':20, 'quali':'nothing'}
fun(**d)
```

```python
def fun(**kwargs):
    for i in kwargs.keys():
        print(i)
d = { 'name':'vish', 'age':20, 'quali':'nothing'}
fun(**d)
```

```python
def fun(**kwargs):
    for i in kwargs.values():
        print(i)
d = { 'name':'vish', 'age':20, 'quali':'nothing'}
fun(**d)
```

## IF is independent Key
## else and elif are dependent key

```python
# if Statement

if True:
    print('Hello')
```

```python
x = input("Enter any value : ")

if x:
    print("Hello")

print('Thanks for visit')
```

```python
# If else

x = input("Enter any value : ")

if x:
    print("Hello")
else:
    print("Thanks for visit")
```

```python
# if-elif

n = int(input("Enter your number : "))

if (n>0):
    print("Positive number")
elif (n<0):
    print("Negative number")
elif (n==0):
    print("zero")
```

```python
# if-elif-else

n = int(input("Enter your number : "))

if (n>0):
    print("Positive number")
elif (n<0):
    print("Negative number")
else:
    print("zero")
```

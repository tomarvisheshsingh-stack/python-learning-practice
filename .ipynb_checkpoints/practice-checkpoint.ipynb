```python
# reverse a given number using a while loop.

num = int(input("Enter the number : "))
revnum = 0
while num > 0:
    a = num % 10
    revnum += a
    revnum = revnum * 10
    num = num//10
rev = revnum // 10

print(rev)

# or
num = int(input("Enter the number: "))
revnum = 0

while num > 0:
    a = num % 10
    revnum = revnum * 10 + a
    num = num // 10

print(revnum)
```

```python
# Check whether a number is a palindrome (e.g., 121).

num = int(input("Enter the number: "))
revnum = 0
n = num
num = abs(num)
while num > 0:
    a = num % 10
    revnum = revnum * 10 + a
    num = num // 10

if n > 0:
    if revnum == n:
        print(n, 'is palindrome')
    else:
        print(n, 'is not palindrome')
else:
    if -revnum == n:
        print(n, "is palindrome")
    else:
        print(n, 'is not palindrome')
```

```python
# Find the factorial of a number using while.
n = int(input("Enter the number you want factorial of : "))
fact = 1
i = 1
while i<=n:
    fact = fact * i
    i += 1
print(f'The factorial of {n} is {fact}')
```

```python
# Print Fibonacci series up to n terms.

n = int(input('Enter no. till where you want to print fibonacci series : '))
a = 0
b = 1
count = 0
while count < n:
    print(a)
    c = a+b
    a = b
    b = c
    count += 1
```

```python
# Find the largest digit in a number.

n = int(input("Enter the number : "))
largest_digit = 0
while n > 0:
    digit = n%10
    if digit > largest_digit:
        largest_digit = digit
    n = n//10
print(largest_digit)
```

```python
# Count how many even and odd digits are in a number.

n = int(input("Enter the number : "))
even_count = 0
odd_count = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even_count += 1
    elif digit % 2 == 1:
        odd_count += 1
    n = n // 10
print(f'number of even digits are {even_count}')
print(f'number of odd digits are {odd_count}')
```

```python
# Print numbers from 1 to 20 but stop the loop if number becomes 15.

num = 1
while num <= 20:
    if num == 15:
        break
    print(num)
    num += 1
```

```python
# Print numbers from 1 to 10 but skip 5

num = 1
while num <= 10:
    if num == 5:
        num += 1
        continue
    print(num)
    num += 1
```

```python
# checking for armstrong no.

n = int(input('Enter any number : '))
x = n
b = x
td = 0
while n>0:
    td = td + 1 
    n = n//10
s = 0
while x>0:
    a = x%10
    s = s + a**td
    x = x//10
if s==b:
    print(f'The number {b} is an armstrong number.')
else:
    print(f'The number {b} is not an armstrong number.')
```

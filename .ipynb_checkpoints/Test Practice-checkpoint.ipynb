```python
##### 1. WAP to print below menetion pattern.
2 # 1 2 3 4 5
3 # 1 2 3 4 5
4 # 1 2 3 4 5
5 # 1 2 3 4 5
6 # 1 2 3 4 5

n = int(input('Enter any no.: '))
for i in range(1, n+1):
    for j in range(1, n+1):
        print(j, end = ' ')
    print()
```

```python
# 2. WAP to print below menetion pattern.
9 # 1
10 # 1 2
11 # 1 2 3
12 # 1 2 3 4
13 # 1 2 3 4 5

n = int(input('Enter any number : '))
for i in range(1, n+1):
    x = 1
    for j in range(1,i+1):
        print(x, end = ' ')
        x += 1
    print()
```

```python
# WAP to print below menetion pattern.
# 16 # 1
# 17 # 2 3
# 18 # 4 5 6
# 19 # 7 8 9 10
a = 1
n = int(input('Enter any number : '))
for i in range(1, n+1):
    for j in range(1,i+1):
        print(a, end=' ')
        a = a+1
    print()
```

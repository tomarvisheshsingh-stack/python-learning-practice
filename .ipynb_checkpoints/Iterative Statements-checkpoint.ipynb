```python
# for checking whether the no. is prime no. or not

num = int(input("Enter the num you want to check : "))
i = 2
count = 0
while i < num:
    if num % i == 0:
        count += 1
    i = i+1
if count == 0:
    print(num, "is a prime number")
    print(f'{num} is a prime number')
else:
    print(num, "is not a prime number")
```

```python
n = int(input('Enter any number : '))
x = n
rev = 0
while n>0:
    a = n%10
    rev = rev*10+a
    n = n//10
if rev==x:
    print(x, 'is a palindrome.')
else:
    print(x, 'is not a palindrome')
```

```python
# armstrong no.
n = int(input('Enter any number : '))
x = n
z = n
s = 0
count = 0
while n>0:
    count = count + 1
    n = n//10
while x>0:
    a = x%10
    s = s + a**count
    x = x//10
if s == z:
    print(z, 'is an armstrong no.')
else:
    print(z, 'is not an armstrong no.')
```

```python
# printing factorial of natural numbers upto n
n = int(input('Enter any number : '))
fact = 1
for i in range(1, n+1):
    if i<n:
        print(i, end = 'x')
    else:
        print(i, end = '=')
    fact = fact * i
print(fact)
```

```python
# fibonacci series

n  = int(input('Enter any number : '))
a = 0
b = 1
while n>0:
    if n>1:
        print(a, end = ', ')
    else:
        print(a)
    z = a+b
    a = b
    b = z
    n = n-1
    
```

## Transfer statements

### continue : skip current iteration
### pass : do nothing in that code block(skip current block)
### break : Terminate any loop

```python
# continue
n = int(input('Enter any value : '))
i = 1
while i<=n:
    if i==3 or i==6 or i==9:
        continue
    print(i)
    i = i + 1
```

```python
n = int(input('Enter any value :'))
i = 1
while i<=n:
    if i==3 or i==6 or i==9:
        i = i + 1
        continue
    print(i)
    i = i + 1
```

```python
# Pass

n = int(input('Enter any value : '))
i = 1
while i<=n:
    if i==3 or i==6 or i==9:
        pass
    print(i)
    i = i + 1
```

```python
# break

n = int(input('Enter any value : '))
i = 1
while i<=n:
    if i==5:
        i = i + 1
        break
    print(i)
    i = i + 1
print('something')
```

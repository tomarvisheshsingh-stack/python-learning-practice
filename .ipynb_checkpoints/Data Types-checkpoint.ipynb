```python
# Index and Find

s = 'this is python'
print(s.index('is'))
print(s.index('is',3))
# print(s.index('si'))
print(s.find('is', 4))         # Find syntax : object.find("Element", start, stop)
print(s.find('si'))
```

```python
# Join 
# join syntax : "separator".join(object)

l = ['python', 'java', 'php']
s = " ".join(l)
print(s)

s1, s2, s3 = "Python", "Java", "PHP"
# a = " ".join(s1, s2, s3)
a = " ".join([s1,s2,s3])
print(a)
print(type(a))

```

```python
s1, s2, s3 = "Python", "Java", "PHP"
# a = " ".join(s1, s2, s3)
a = ", ".join([s1,s2,s3])
print(a)
```

```python
# Split 
# split syntax = string.split("separator", no. of splits)      by default: separator takes space and no. of splits takes no. of spaces
# It also replaces the separator: which means the separator doesn't show up in the output
s = "python java php"
l = s.split()
a = s.split(" ", 2)
b = s.split("p")
print(b)
print(l)
print(a)
```

## list methods

### append() : used to single element at last position 
### copy() : copies the list
### clear() : empties the list
### extend() : adding multiple elements at end on list
### insert() : add single element at desired position
### pop() : deletes last element from list
### remove() : delete targeted element from the list(first occurenece)
### index() : gives indexing value of the elements in list
### count() : counts no. elements in the list 
### sort() : arrange lists in to ascending order based on ascii values
### reverse() : reverses the order of list

```python
# append syntax : list.append(element to be added) 

l = [10, 20, 65, 5, 'python', 'ans']
l.append('java')
print(l)
```

```python
# copy syntax : list2 = list.copy()

l = [10, 20, 65, 5, 'python', 'ans']
l1 = l.copy()
print(l, l1)
print(id(l), id(l1))
```

```python
# clear syntax : list.clear()

l = [10, 20, 65, 5, 'python', 'ans']
print(l)
l.clear()
print(l)
print(id(l))

# del l         # del is used to delete from memory
print(l)
```

```python
# extend syntax : list.extend(elements)

l = [10, 20, 65, 5, 'python', 'ans']
l.extend('java')
l.extend(['Java', 'php'])
print(l)
```

```python
# insert syntax : list.insert( position(indexing value), 'element')

l = [10, 20, 65, 5, 'python', 'ans']
l.insert(2, "java")
print(l)
l.insert(1, [1,7,8,6])
print(l)
```

```python
# pop syntax : list.pop(indexing no.)

l = [10, 20, 65, 5, 'python', 'ans']
l.pop()
l.pop(1)
print(l)
```

```python
# remove syntax : list.remove(element)

l = [10, 20, 65, 5, 'python', 'ans']
l.remove('python')
print(l)
l.remove(20)
print(l)
```

```python
# index syntax : list.index(element, start, stop)

l = [10, 20, 65, 5, 'python', 'ans']
print(l.index(20))
```

```python
# count syntax : list.count()
l = [10, 20, 65, 10, 5, 'python', 'ans']
print(l.count(10))
```

```python
# sort syntax : list.sort()

l = ['python', 'ans']
l.sort()
print(l)
l2 = [10, 20, 65, 5]
l2.sort()
print(l2)
```

```python
# reverse syntax : list.reverse()

l2 = [10, 20, 65, 5]
l2.reverse()
print(l2)

l = [10, 20, 65, 5, 'python', 'ans']
l.reverse()
print(l)

# to arrange any list in descending first sort the list than reverse it
l2.sort(reverse = True)
print(l2)
```

```python
# print syntax : print( "printable", end = '\n', sep = " ")

print('hello', 'world', sep = ',', end = " ")
print('bye')
```

# Dictionary Methods

### copy() : copies the dictionary
### clear() : clears the dictionary
### get() : used to fetch values on basis of keys
### pop() : used to remove key value pair based on key
### popitems() : deletes the last item by default
### keys() : used to return all keys of dictionary
### values() : used to return all values of dictionary
### items() : used to return all items of dictionary
### update() : used to add a dictionary to another dictionary
### set_default() : used to add a item to dictionary with key value pair(if the key already exists in dictionary no changes are made, and returns the existing value of that key)


```python
# copy()
d = {'x': 10, 'b' : 20, "c":30}
d1 = d.copy()
print(d1)
print(id(d), id(d1))
print()

# clear()
d1.clear()
print(d1)
```

```python
# get()
# syntax : d.get('keys', 'default value')

d = {'x': 10, 'b' : 20, "c":30}
print(d.get('x'))
print(d.get('a', 'no value for it'))
print(d['x'])
print(d)
d['x'] = 100
# d.get('b') = 30
print(d)
# print(d['a'])
```

```python
# pop()
# syntax : d.pop('key')

d = {'x': 'ab', 'b' : 20, "c":30}
print(d.pop('x'))
# d.pop()
print(d)
```

```python
# popitem()
# syntax : d.popitem()

d = {'x': 'ab', 'b' : 20, "c":30}
print(d.popitem())
print(d)
```

```python
# keys 
# syntax : d.keys()

d = {'x': 'ab', 'b' : 20, "c":30}
print(d.keys())
```

```python
# values
# syntax : d.values()

d = {'x': 'ab', 'b' : 20, "c":30}
print(d.values())
```

```python
# items 
# syntax : d.items()
d = {'x': 'ab', 'b' : 20, "c":30}
print(d.items())
```

```python
# update
# syntax : d.update(d1)

d = {'x': 'ab', 'b' : 20, "c":30}
d1 = {1:45, 5:87}

d.update(d1)
print(d)
```

```python
# set_default
# syntax : d.setdefault()

d = {'x': 'ab', 'b' : 20, "c":30}
print(d.setdefault('d', 'ac'))
print(d.setdefault('d', 'aa'))
print(d)
```

## Set Methods

```python
# copy() : uses to copy set into another

s = {10,20,10,30,'python'}
a = s.copy()
print(a)
print(id(s), id(a))

# clear() : uses to empty a set
a.clear()
print(a)
```

```python
# add : adds an element to the set

s = {10,20,10,30,'python'}
s.add('num')
print(s)

# update() : used to add multiple elements
s.update({45,56,45})
s.update('python')
print(s)
```

```python
# remove : used to remove a element
s = {10,20,10,30,'python'}
# s.remove(105)
print(s)

# discard : removes element but doesn't throw a error
s = {10,20,10,30,'python'}
s.discard(50)
print(s)
```

```python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8, 9}

# union : used to join two sets
s = s1.union(s2)
print(s)
print()

# intersections : gives the common value of two sets
ss = s1.intersection(s2)
print(ss)
print()

# difference : gives all the element of set 1 that are not in set 2
print(s1.difference(s2))
print(s2.difference(s1))
print()

# symmetric_difference() : gives all the values of both sets except the common values
print(s1.symmetric_difference(s2))
print()

# intersection_update() : same as intersection but updates the existing set only
print(s1.intersection_update(s2))
print(s1)
print()

# difference_update : same as difference but updates the existing set only
print(s1.difference_update(s2))
print(s1)
print()

# symmetric_difference_update : same as symmetric_difference but updates the existing set only
print(s1.symmetric_difference_update(s2))
print(s1)
```

```python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8, 9}

# isdisjoint : checks whether the sets are joined or not
print(s1.isdisjoint(s2))
print()

# issubset : checks whether the set 1 is subset of set 2
print(s1.issubset(s2))
print()

# issuperset : checks whether the set 1 is superset of set 2
print(s1.issuperset(s2))
```

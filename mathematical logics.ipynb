```python
# factors of a number
n = int(input('Enter any number : '))
for i in range(1, n+1):
    if n % i == 0:
        if i<n:
            print(i, end = ', ')
        else:
            print(i)
```

```python
# factorial of number 
n = int(input('Enter any number : '))
fact = 1
for i in range(1, n+1):
    fact = fact * i
    if i < n:
        print(i, end = 'x')
    else:
        print(i, end = '=')
print(fact)
```

```python
# armstrong number
n = int(input('Enter any value : '))
a = n
c = n
count = 0
while n>0:
    n = n // 10
    count = count + 1
s = 0
while a>0:
    b = a%10
    s = s + b**count
    a = a//10
if s == c:
    print(f'{c} is an armstrong number.')
else:
    print(f'{c} is not an armstrong number.')
```

```python
# palindrome number
n = int(input('Enter any number : '))
an = abs(n)
rev = 0
while an>0:
    r = an%10
    rev = rev*10 + r
    an = an//10
if n>0:
    if n == rev:
        print(f'{n} is a palindrome.')
    else:
        print(f'{n} is not a palindrome.')
else:
    rev = rev - 2*rev
    if rev == n:
        print(f'{n} is a palindrome.')
    else:
        print(f'{n} is not a palindrome.')
```

```python
# hcf of a number 
n1 = int(input('Enter any number : '))
n2 = int(input('Enter any number : '))

small = min(n1, n2)
for i in range(2, small+1):
    if n1 % i == 0 and n2 % i == 0:
        hcf = i
print(f'H.C.F. is {hcf}')
```

```python
# lcm of a number
n1 = int(input('Enter any number : '))
n2 = int(input('Enter any number : '))

small = min(n1, n2)
for i in range(2, small+1):
    if n1 % i == 0 and n2 % i == 0:
        hcf = i

lcm = (n1*n2)/hcf
print(f'L.C.M of {n1} and {n2} is {lcm}')
```

```python
# fibonacci series upto n 
n = int(input('Enter any number : '))
a = 0
b = 1
i = 1
while i<=n:
    if i<n:
        print(a, end = ', ')
    else:
        print(a)
    x = a+b
    a = b
    b = x
    i = i+1
```

```python
# harshad number
n = int(input('Enter any value : '))
s = 0
x = n
while n>0:
    a =  n%10
    s = s+a
    n = n // 10
if x % s == 0:
    print(f'{x} is a harshad number.')
else:
    print(f'{x} is not a harshad number.')
```

```python
# neon numbers
n = int(input('Enter any number : '))
s = n**2
sum = 0
while s>0:
    a = s%10
    s = s//10
    sum = sum + a
if sum == n:
    print(f'{n} is a neon number.')
else:
    print(f'{n} is not a neon number.')
```

```python
# peterson number
n = int(input('Enter any number : '))
x = n
sum = 0
while n>0:
    a = n%10
    n = n // 10
    fact = 1
    for i in range(1, a+1):
        f = 0
        fact = fact*i
        i = i+1
        if f<fact:
            f = fact
    sum = sum + f
if sum == x:
    print(f'{x} is a peterson number.')
else:
    print(f'{x} is not a peterson number.')
```

```python
# spy numbers
n = int(input('Enter any number : '))
x = n
sum = 0
pro = 1
while n>0:
    a = n%10
    sum = sum + a
    pro = pro * a
    n = n//10
print(sum)
print(pro)
if sum == pro:
    print(f'{x} is a spy number.')
else:
    print(f'{x} is not a spy number.')
```

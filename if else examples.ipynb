```python
# to check whether the given number is +ve, -ve or 0
n = float(input('Enter any number : '))
if n>0:
    print(f'{n} is a positve number.')
elif n==0:
    print(f'{n} is zero.')
else:
    print(f'{n} is a negative number.')
```

```python
# to check the greater number between three numbers
n1 = float(input('Enter any number : '))
n2 = float(input('Enter any number : '))
n3 = float(input('Enter any number : '))
if n1 >= n2:
    if n1 > n3:
        print(f'{n1} is greatest among these three.')
    else:
        print(f'{n3} is greatest among these three.')
elif n2 >= n1:
    if n2 > n3:
        print(f'{n2} is greatest among these three.')
    else:
        print(f'{n3} is greatest among these three.')
```

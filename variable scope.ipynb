```python
x = 10
def new():
    x = 20
    print(x)
new()
print(x)
```

```python
# converting local variable into global variable
x = 10
def new():
    global x
    x = 20
    print(x)
new()
print(x)
```

```python
# modifying global variable by local variable
x = 10
def new():
    global x
    x = 20
print(x)
new()
print(x)
```

```python
def new():
    x = 10
    def new1():
        x = 20       # nonlocal variable
        print(x)
    new1()
    print(x)
new()
```

```python
def new():
    x = 10
    def new1():
        def new2():
            nonlocal x
            x = 20
            print(x)
        new2()
        print(x)
    new1()
    print(x)
new()
```

```python
x = 10
def new():
    x = 20
    print('local variable: ', x)
    print('global variable:', globals()['x'])
new()
```

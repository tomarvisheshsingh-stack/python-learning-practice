```python
# Question 1

n = int(input('Enter any number : '))

if n > 0:
    if n%2==0:
        print(n,'is positive and even.')
    else:
        print(n,'is positive and odd.')
elif n < 0:
    print(n,'is negative.')
else:
    print(n,'is zero.')
```

```python
# Question 3
h = int(input('Enter you marks (out of 100) in hindi : '))
m = int(input('Enter you marks (out of 100) in maths : '))
e = int(input('Enter you marks (out of 100) in english : '))
s = int(input('Enter you marks (out of 100) in science : '))
ss = int(input('Enter you marks (out of 100) in social science : '))

total = h+e+m+s+ss
per = (total/500)*100
print(per)
if per>=65:
    print('You got A Division.')
elif per >= 50:
    print('You got B Division.')
elif per >= 33:
    print('You got C Division.')
else:
    print("You've failed the class.")
```

```python
# Question 4
s1 = int(input('Enter the length of side 1 : '))
s2 = int(input('Enter the length of side 2 : '))
s3 = int(input('Enter the length of side 3 : '))

if s1 == s2 == s3:
    print("The triangle is an equilateral triangle.")
elif s1 == s2 != s3:
    print("The triangle is an isosceles triangle.")
elif s1 == s3 != s2:
    print("The triangle is an isosceles triangle.")
elif s1 != s2 == s3:
    print("The triangle is an isosceles triangle.")
else:
    print("The triangle is a scalene triangle.")
```

```python
# Question 5
n = int(input('Enter any number : '))
su_ev = 0
su_od = 0
cou_ev = 0
cou_od = 0
for i in range(1, n+1):
    if i%2 == 0:
        su_ev +=  i
        cou_ev += 1
    else:
        su_od += i
        cou_od += 1
print(f'The sum of all even number till {n} is {su_ev}\nThe count of all even number till {n} is {cou_ev}')
print(f'The sum of all odd number till {n} is {su_od}\nThe count of all odd number till {n} is {cou_od}')
```

```python
# Question 6
n = int(input('Enter the number you want to check is prime or not : '))
count = 0
for i in range(2, n):
    if n%i==0:
        count += 1

if n == 1:
    print('Invalid input: 1 is neither prime nor non-prime')
elif count == 0:
    print(n,'is a prime number.')
else:
    print(n,'is not a prime number.')
```

```python
# question 9
n = int(input('Enter the point till where you want fibonacci series : '))
a = 0
b = 1
while n>0:
    print(a, end = ' ')
    c = a+b
    a = b
    b = c
    n = n-1
```
